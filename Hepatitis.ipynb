#7) Menentukan Label

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
X= data_clean.drop('Class', axis=1)
y=data_clean['Class']
```

```python
#Memvisualisasikan kelas target
data_clean['Class'].value_counts().plot(kind='bar',figsize=(8,5),color=['green','Blue'])
plt.title("Menampilkan class target")
plt.show()
```

```python
#Menyeimabngkan data dengan menggunakan smote
smote=SMOTE(random_state=42)
x_smote,y_smote=smote.fit_resample(X,y)
```

```python
#Membagi data dengan menggunakan train test split dengan pembagian data 20 % untuk data uji dan 80% train
X_train,X_test,y_train,y_test=train_test_split(x_smote,y_smote, test_size=0.3, random_state=42, stratify=y_smote)
```

#8) Membangun model

```python
# import library untuk membangun model
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
rf= RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
nb= GaussianNB()
nb.fit(X_train,y_train)
```

#9) Mengevaluasi Model

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
y_pred_dt=dt.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred_dt))
print("Laporan Klasifikasi dari Model Decision Tree")
print(classification_report(y_test,y_pred_dt))
```

```python
cm_dt=confusion_matrix(y_test,y_pred_dt)
sns.heatmap(cm_dt, annot=True,fmt='d',cmap='Blues', xticklabels=['Live','Die'], yticklabels=['live','Die'])
plt.xlabel("Predict")
plt.ylabel("Actual")
plt.title("Confussion Matriks Dari pemodelan Decision Tree")
plt.show()
```

```python
y_pred_rf = rf.predict(X_test)
print("Accuracy score: ",accuracy_score(y_test,y_pred_rf))
print("Laporan klasifikasi Random Forest")
print(classification_report(y_test,y_pred_rf))
```

```python
cm_rf=confusion_matrix(y_test,y_pred_rf)
sns.heatmap(cm_rf,annot=True, cmap='Blues', fmt='d',xticklabels=['Live','Die'],yticklabels=['Live','Die'])
plt.xlabel("Predict")
plt.ylabel("Actual")
plt.title("Confussion Matriks dari Random Forest")
plt.show()
```

```python
y_pred_lr=lr.predict(X_test)
print("Accuracy score: ",accuracy_score(y_test,y_pred_lr))
print("Laporan Klasifikasi dari Logistik Regression")
print(classification_report(y_test,y_pred_lr))
```

```python
cm_lr=confusion_matrix(y_test,y_pred_lr)
sns.heatmap(cm_lr,annot=True, cmap="Blues", fmt='d', xticklabels=['Live','Die'],yticklabels=['Live','Die'])
plt.xlabel("Predict")
plt.ylabel("Actual")
plt.title("Confusion Matrik dari Logistic Regression")
plt.show()
```
